```python
from os import listdir
from os.path import isfile, join
```

```python
onlyfiles = [f for f in listdir("./proj/data/searching_for_science/") if isfile(join("./proj/data/searching_for_science/", f))]
```

```python
onlyfiles
```

```python
len(onlyfiles)
```

```python
t = [file if file[-4:] == "html" else None for file in onlyfiles] 
onlyfiles = [i for i in t if i]
onlyfiles[3]
len(onlyfiles)
```

```python
for file in onlyfiles[6::]:
    splited = file.split("_")
    print(splited)
    params = {
      "q": splited[1],
      "device": "desktop",
      "hl": "iw" if splited[2].lower() == "hw" else splited[2].lower(),
      "gl": splited[3].split(".")[0].lower(),
      "google_domain": "google.com",
      "api_key": "63cbd78f9490636a3a8fa5f5692b58d7d965a717dd9636d0e68b881eae5cb472"
    }
    search = GoogleSearch(params)
    results = search.get_dict()
    json_object = json.dumps(results, indent = 4, ensure_ascii=False)
    file_name = file[:-5] + ".json"
    jsonFile = open("./proj/data/searching_for_science/" + file_name, "w")
    jsonFile.write(json_object)
    jsonFile.close()
    
```
